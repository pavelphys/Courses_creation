## Функции

Хотя заманчиво работать только со сценариями, представляющими собой файл, полный инструкций, следующих друг за другом, все же лучше разбивать код на отдельные структурные единицы. Для этого можно использовать функции.
Их можно использовать повторно в коде, в отличие от скрипта. При этом функциональное мышление помогает разбить проблему на мелкие блоки.

Функции могут ничего не принимать в качестве аргумента, но вне зависимости от реализации они будут возвращать какой-либо объект после выполнения тела функции. В простейшем случае результатом действия функции будет специальный объект типа None. По существу это просто команда вернуть "Ничего".

````python
def my_function1():
    """
    Функция ничего не делает
    """


def my_function2():
    """
    Функция тоже ничего не делает. 
    Оператор pass может использоваться когда синтаксически
    требуется присутствие оператора, но от программы не требуется действий.  
    """   
    pass 


def my_function3():
    return


def my_function4():
    return None
````

````python
print(my_function1(), my_function2(), my_function3(), my_function4())
````

Отступы в Python обязательны! Каждый командный блок, следующий за двоеточием, имеет дополнительный уровень отступа по отношению к предыдущей строке с двоеточием. 
Строгое соблюдение отступов - это плата за избавление от символов **{** или **;** выделяющих логические блоки в других языках. Неправильная расстановка отступов приводит к ошибкам!

Напишем функцию возведения числа в квадрат. Мы можем явно указать, объект какого типа мы хотим передавать в качестве аргумента, но это делать вовсе не обязательно т. к. интерпретатор сам определит что мы кладем в функцию.

````python
def square(x:int):
    """
    Возвращает квадрат числа x.
    """
    return x**2

print(square(10))    
````

Функции могут содержать произвольное число аргументов. Часть из них могут иметь значение по умолчанию. Это значит, что если мы не указали аргумент при вызове, он будет взят из значения по умолчанию.

````python
def linear_shift(x, a=10, b=-5):
    return a * x + b

print(linear_shift(1))
print(linear_shift(1, 1))
print(linear_shift(1, 1, 2))
````

## Условные конструкции

Оператор if используется для проверки условий: если условие верно, выполняется блок выражений (называемый «if-блок»), иначе выполняется другой блок выражений («else-блок»). 
Блок «else» является необязательным.

Итак, условная инструкция в Питоне имеет следующий синтаксис:


```
if Условие:
    Блок инструкций 1
else:
    Блок инструкций 2
```

Блок инструкций 1 будет выполнен, если Условие истинно. Если Условие ложно, будет выполнен Блок инструкций 2.

````python
# Вывести модуль
x = int(input('Введите число '))
if x >= 0:
    print('Его модуль равен', x)
else:
    print('Его модуль равен',-x)
````

````python
print(abs(x))
````

## Синтаксис более сложных условных конструкций:

Специальная команда elif эквивалентна вызову if сразу после команды else. Таким образом можно упростить запись.

```
if Условие_1:
    Блок инструкций_1
else:
    if Условие_2:
        Блок инструкций_2
    else:
        Блок инструкций_3

```
Эквивалентно:


```
if Условие_1:
    Блок инструкций_1
elif: Условие_2:
    Блок инструкций_2
else:
    Блок инструкций_3
```





````python
# Попадание точки в интервал
x = int(input('Введите число '))

if 0 < x < 10:
    print("Точка попала в интервал (0, 10)")

elif 20 < x < 30:
    print("Точка попала в интервал (20, 30)")

else:
    print("Точка не попала ни в один из двух интервалов")
````

# Циклы

В цикле for указывается переменная и множество значений, по которому будет пробегать переменная. Множество значений может быть задано списком, кортежем, строкой или диапазоном.

Как правило, циклы for используются либо для повторения какой-либо последовательности действий заданное число раз, либо для изменения значения переменной в цикле от некоторого начального значения до некоторого конечного.

````python
start = 0
stop = 10
step = 1

for i in range(start, stop, step):
    print(i)
````

Для повторения цикла некоторое заданное число раз n можно использовать цикл for вместе с функцией range(n)

````python
for i in range(10):
    print(i)
````

````python
for i in range(stop, start, -step):
    print(i)
````

Оператор while позволяет многократно выполнять блок команд до тех пор, пока выполняется некоторое условие. Это один из так называемых операторов цикла. Он также может иметь необязательный пункт else.

````python
while(start < stop):
    print(start)
    start += step
else:
    print("start >= stop")
````

````python
while(True):
    string = input()
    if string == 'конец':
        break
    elif string == 'почти':
        continue
    else:
        print("еще не конец")
````

# Списки

Список представляет собой пронумерованную последовательность элементов. Список можно задать перечислением элементов в квадратных скобках. Индексы элементов отсчитываюся от нуля. То есть первому элементу списка соответствует индекс 0. Обращаясь к списку по индексу можно получить значение соответствующего элемента. Быстро получить последний элемент списка можно обратившись по индексу -1. Предпоследнему будет соответствовать индекс -2 и так далее.

````python
x = [0, 1, 2, 3, 4, 5, 6, 7, 8]

first_element, *rest = x
print(first_element)
print(rest)
````

Со списками, так же как и со строками, можно делать срезы. А именно:

A[i:j]  срез из j-i элементов A[i], A[i+1], ..., A[j-1].

A[i:j:-1]  срез из i-j элементов A[i], A[i-1], ..., A[j+1] (то есть меняется порядок элементов).

A[i:j:k]  срез с шагом k: A[i], A[i+k], A[i+2*k],... . Если значение k<0, то элементы идут в противоположном порядке.

Каждое из чисел i или j может отсутствовать, что означает “начало строки” или “конец строки”

Списки, в отличии от строк, являются изменяемыми объектами: можно отдельному элементу списка присвоить новое значение. Но можно менять и целиком срезы.

````python
print(x[0])
print(x[-1])
print(x[::-1])
print(x[0:3])
````

Со списками можно легко делать много разных операций:

*   x in A	Проверить, содержится ли элемент в списке. Возвращает True или False
*   x not in A	То же самое, что not(x in A)
*   min(A)	Наименьший элемент списка
*   max(A)	Наибольший элемент списка
*   A.index(x)	Индекс первого вхождения элемента x в список, при его отсутствии генерирует исключение ValueError
*   A.count(x)	Количество вхождений элемента x в список

````python
z1 = list(map(square, x))
print(z1)
print(max(z1))
````

# Построение графиков

````python
# Создание графика и координатных осей 
fig, ax = plt.subplots(figsize=(8, 6))


# Область обображения графика
ax.set_xlim(0, 8)
ax.set_ylim(0, 65)


# Построение графика 
plt.plot(x, y, color="blue", linewidth=3, label="Парабола")

# Подпись к графику
ax.set_title("Название графика", pad=10, fontsize=20)

# Обозначения для координатных осей
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)

# Отображение легенды графика
plt.legend(loc='upper left', fontsize=15)

# Побочные деления на осях
ax.minorticks_on()

# Сетка
plt.grid(which='major',linestyle=':', linewidth=2, color="grey", alpha=0.95)

# Выполнить сценарий отображения
plt.show()
````
